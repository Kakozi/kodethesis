```python
import pandas as pd
```

```python
df = pd.read_csv("IV\IV/2436_36_1_03272019.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(df['V'], df['I'], 'b-', linewidth=2)

plt.title('IV Curve')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.grid(True)
plt.show()
```

```python
df['P'] = df['V'] * df['I']

fig, ax1 = plt.subplots(figsize=(8,6))

# IV curve
ax1.plot(df['V'], df['I'], 'b-', label='IV Curve')
ax1.set_xlabel('Voltage (V)')
ax1.set_ylabel('Current (A)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# PV curve
ax2 = ax1.twinx()
ax2.plot(df['V'], df['P'], 'r--', label='PV Curve')
ax2.set_ylabel('Power (W)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

fig.suptitle('IV and PV Curves')
fig.tight_layout()
plt.show()
```
