```python
import pandas as pd
meta = pd.read_csv('AnonDB.csv')
```

```python
import os
from PIL import Image
from tqdm import tqdm

# === Configuration ===
input_folder = "EL/EL_wrp/"
output_folder = "EL_resampled/"
target_size = (512, 512)

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# === Collect all filenames from both columns ===
all_files = list(meta['ELHPath'].dropna()) + list(meta['ELLPath'].dropna())

# Clean and keep only filenames
all_files = [os.path.basename(str(f)) for f in all_files if str(f).lower().endswith('.tiff')]
all_files = list(set(all_files))  # remove duplicates

# === Loop through each file ===
for filename in tqdm(all_files, desc="Resampling EL images"):
    try:
        input_path = "EL/" + filename              # ✅ simply add 'EL/' prefix
        output_path = "EL_resampled/" + filename

        with Image.open(input_path) as img:
            img_resized = img.resize(target_size, Image.LANCZOS)
            img_resized.save(output_path)

    except Exception as e:
        print(f"⚠️ Error processing {input_path}: {e}")

print("✅ Resampling completed successfully!")
print(f"Saved to: {os.path.abspath(output_folder)}")

```
